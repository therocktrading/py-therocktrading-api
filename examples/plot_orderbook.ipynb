```python
import plotly.graph_objs as go
import pandas as pd

df = pd.read_csv("data/orderbook.csv", index_col=False)
df.head()
```

```python
df=df[df["pairs"] == "SACUSDT"]
```

```python

trace1 = {
  "mode": "lines+markers", 
  "name": "bid", 
  "type": "scatter", 
  "x": df.price[df.orderbook == "bid"],
  "y": df.amounts[df.orderbook == "bid"]
}
trace2 = {
  "mode": "lines+markers", 
  "name": "ask", 
  "type": "scatter", 
  "x": df.price[df.orderbook == "ask"],
  "y": df.amounts[df.orderbook == "ask"]
}
data = ([trace1, trace2])
layout = {
  "title": "Limited Order Book", 
  "xaxis": {"title": "price"}, 
  "yaxis": {"title": "amount"}
}
fig = go.Figure(data=data, layout=layout)
fig.write_html("orderbook.html")
```
